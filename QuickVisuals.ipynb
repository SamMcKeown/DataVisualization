## Section 1

#### Trends - A pattern of change over a period of time
- Line charts are best to show trends over a period of time & multiple lines can be used to show trends in more than one group

#### Relationship - The relationship between variables in your data
- Bar charts are useful for comparing quantities corresponding to different groups


- Heatmaps are used to find colour-coded patterns in tables of numbers


- Scatterplots show the relationship between two continuous variables. If colour coded we can also show the relationship with a third categotical variable

- Adding a regression line to the scatterplot makes it easier to see any linear relationship between two variables

- An LMPLOT is useful for drawing multiple regression lines, if the scatter plot contains multiple colour-coded groups


- A Swarmplot / Categorical scatterplot shows the relastionship between a continuous variable & a categorical variable

## Section 2

### 1. Line Chart

```python
##
data = pd.read_csv('data/spotify.csv', index_col='Date', parse_dates=True)
display(data.head(3), data.tail(3))
```

```python
## Set the width & height of figure
plt.figure(figsize=(14, 6))

## Title
plt.title('Spotify: Daily Global Streams of Popular Songs in 2017-2018')

## Lineplot
sns.lineplot(data=data)

## Control Scientific Notation
plt.yticks(rotation=0)
plt.ticklabel_format(style='plain', axis='y')
```

#### 1-2. Plot a subset of the data

```python
## Column Names
list(data.columns)
```

```python
plt.figure(figsize=(14, 6))

plt.title('Spotify: Daily Global Streams of Popular Songs in 2017-2018')

## Lineplot 1
sns.lineplot(data=data['Shape of You'], label='Shape of You')

## Lineplot 2
sns.lineplot(data=data['Despacito'], label='Despactio')

## x-label
plt.xlabel('Date')

## y-label
plt.ylabel('Number of Streams')

plt.yticks(rotation=0)
plt.ticklabel_format(style='plain', axis='y');
```

### 2. Bar Chart

```python
##
data = pd.read_csv('data/flight_delays.csv', index_col='Month')
data.head(3)
```

```python
## Set the width & height of figure
plt.figure(figsize=(10, 6))

## Title
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

## Bar chart
sns.barplot(x=data.index, y=data['NK'], ec='black')

## y-label
plt.ylabel('Arrival delay (in minutes)');
```

### 3. Heatmap

```python
plt.figure(figsize=(14, 7))

plt.title('Average Arrival Delay for Each Airline, by Month')

## Heatmap 
sns.heatmap(data=data, annot=True)

## x-label
plt.xlabel('Airline');
```

### 4. Scatter Plot

```python
##
data = pd.read_csv('data/insurance.csv')
data.head(3)
```

```python
plt.figure(figsize=(10, 6))

## Scatterplot
sns.scatterplot(x=data['bmi'], y=data['charges']);
```

#### 4.1 Add a regression line

```python
plt.figure(figsize=(10, 6))

## Reg plot
sns.regplot(x=data['bmi'], y=data['charges']);
```

#### 4.2 Colour-coded scatter plots

```python
plt.figure(figsize=(10, 6))

## Reg plot - how smoking affects the relationship between BMI & insurance costs
sns.scatterplot(x=data['bmi'], y=data['charges'], hue=data['smoker']);
```

#### 4.3 Add two regression lines

```python
plt.figure(figsize=(14, 8))

## lmplot
sns.lmplot(x='bmi', y='charges', hue='smoker', data=data);
```

### 5. Swarmplot - Categorical Scatter Plot

```python
plt.figure(figsize=(10, 6))

sns.swarmplot(x=data['smoker'], y=data['charges']);
```

### 6. Histogram

```python
##
data = pd.read_csv('data/iris.csv', index_col='Id')
data.head(3)
```

```python
plt.figure(figsize=(10, 6))

## Histogram 
sns.histplot(data['Petal Length (cm)']);
```

### 7. Density plots

```python
plt.figure(figsize=(10, 6))

## KDE plot 
sns.kdeplot(data=data['Petal Length (cm)'], shade=True);
```

#### 7.1 2D KDE plots

- the curve at the top of the figure is a KDE plot for the data on the x-axis ( in this case, __iris_data['Petal Length (cm)']__ )


- the curve on the right of the figure is a KDE plot for the data on the y-axis ( in this case, __iris_data['Sepal Width (cm)']__ ).

```python
plt.figure(figsize=(10, 6))

## 2D KDE plot
sns.jointplot(x=data['Petal Length (cm)'], y=data['Sepal Width (cm)'], kind='kde');
```

#### 7.2 Colour-coded plots

```python
plt.figure(figsize=(10, 6))

## Histograms for each species
sns.histplot(data=data, x='Petal Length (cm)', hue='Species')

## Add title
plt.title('Histogram of Petal Lengths, by Species');
```

```python
plt.figure(figsize=(10, 6))

## KDE plots for each species
sns.kdeplot(data=data, x='Petal Length (cm)', hue='Species', shade=True)

## Add title
plt.title('Distribution of Petal Lengths, by Species');
```
